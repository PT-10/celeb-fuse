```python
import os
from pathlib import Path
from diffusers import StableDiffusionPipeline
import torch

output_dir = "/kaggle/working/generated_images_male_non_blond"
os.makedirs(output_dir, exist_ok=True)


model_path = Path("/kaggle/input/male_not_blond/pytorch/default/1/dreambooth_male_not_blond_model")
pipe = StableDiffusionPipeline.from_pretrained(model_path, torch_dtype=torch.float16).to("cuda")

pipe.safety_checker = lambda images, **kwargs: (images, [False] * len(images))


from tqdm import tqdm

prompt = "a photo of a nonblondmankk person"
negative_prompt = "blurry, low quality, deformed"

for i in tqdm(range(5000)):
    image = pipe(prompt, num_inference_steps=30, guidance_scale=7.5, negative_prompt=negative_prompt).images[0]
    image.save(f"{output_dir}/blond_male_{i+1:05}.png")
```
